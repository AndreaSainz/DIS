```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(t, a, b, c):
    return np.sqrt(a) * np.exp(-b * t) * np.sin(c * t) + 0.5 * np.cos(2 * t)

# Example usage
# Define parameters
a = 0.1
b = -0.13
c = 9


# Define time range
t = np.linspace(0, 1, 100)  # time from 0 to 10 with 100 points

# Calculate f(t) for these parameters
f_values = f(t, a, b, c)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(t, f_values, label=r'$f(t ; {a}, {b}, {c})$'.format(a=a, b=b, c=c), color='blue')
plt.xlabel('Time $t$')
plt.ylabel(r'$f(t)$')
plt.title(r'Time Series $f(t ; {a}, {b}, {c})$'.format(a=a, b=b, c=c))
plt.legend()
plt.grid(True)
plt.show()
```

```python
from ipywidgets import interactive
import ipywidgets as widgets


def plot_with_parameters(a, b, c):
    # Define time range
    t = np.linspace(0, 1, 100)

    # Calculate f(t) for these parameters
    f_values = f(t, a, b, c)

    # Plot the function
    plt.figure(figsize=(10, 6))
    plt.plot(t, f_values, label=r'$f(t; a, b, c)$', color='blue')
    plt.xlabel('Time $t$')
    plt.ylabel(r'$f(t)$')
    plt.title(r'Time Series $f(t; a, b, c)$')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create interactive widget
interactive_plot = interactive(
    plot_with_parameters,
    a=widgets.FloatSlider(min=0, max=1, step=0.01, value=0.1),
    b=widgets.FloatSlider(min=-0.5, max=0.5, step=0.01, value=-0.13),
    c=widgets.FloatSlider(min=5, max=10, step=0.1, value=9)
)

display(interactive_plot)
```

```python
# Define parameters
a = 0.1
b = -0.13
c = 9

# Define time range
t = np.linspace(0, 1, 100)  # time from 0 to 10 with 100 points
```

```python
def interpolate_f(x,t, a, b, c):
    y = f(t, a, b, c) #Call the function defined in cell 1 with those parameters fixed
    y_new = np.interp(x, t, y) #Interpolate the function at the points x
    return y_new
```

# Question 2 and 3


```python
x = np.linspace(0, 1, 200)
y_new = interpolate_f(x, t, a, b, c)
y = f(x, a, b, c)

plt.figure()
plt.plot(x, y, label='Real function f(x)',  marker='o' ,color='red', markersize=5)
plt.plot(x, y_new, label='Interpolated points', marker='o', color='blue', markersize=1)
plt.legend()
plt.grid(True)
plt.show()
```

# Question 4

```python
# Compute the ratio
ratio = y/y_new

plt.figure(figsize=(15,8))
plt.plot(x, y, label='Real function scalated f(x)' ,color='red')
plt.scatter(x, ratio, label='Ratio between true values and interpolated values', marker='o', color='green', s= 8)
plt.legend()
plt.grid(True)
plt.show()
```

```python
def interpolate_a(a_new, a, t, b, c ):
    # Create a dataframe
    df = pd.DataFrame(index = t)

    for i in a:
        df['f({i:.3f},b,c)'.format(i=i)] = f(t, i, b, c)
    
    
    # Create new DataFrame for intrpolated values
    df_interpolated = pd.DataFrame(index=t)

    # Interpolate for each value in  'a_new'
    for i_new in a_new:
        # Find the a values thta are closer to  i_new 
        i_left = max([i for i in a if i <= i_new], default=a[0])
        i_right = min([i for i in a if i >= i_new], default=a[-1]) # a[-1] last value of a list

        # Column names so they are all the same size
        left_col = f'f({i_left:.3f},b,c)'
        right_col = f'f({i_right:.3f},b,c)'

        # Interpolation between two columns
        if left_col == right_col:
            # If i_new equal to a value in a, then we copu the column 
            df_interpolated[f'f({i_new:.3f},b,c)'] = df[left_col]
        else:
            # Do interpolation
            interpolated_column = []
            for t_val in t:
                left_val = df[left_col].loc[t_val]
                right_val = df[right_col].loc[t_val]
                interp_val = np.interp(i_new, [i_left, i_right], [left_val, right_val])
                interpolated_column.append(interp_val)

            # Add interpolated column to the dataFrame
            df_interpolated[f'f({i_new:.3f},b,c)'] = interpolated_column

    return df_interpolated
```

```python
# Define a range
a = np.linspace(0, 1, 10)
# Define time range
t = np.linspace(0, 1, 100)

a_new = np.linspace(0, 1, 20).tolist()
a_new.append(0.125)
a_new.sort()
```

```python
df_inter = interpolate_a(a_new, a, t, b, c )
```

```python
plt.figure(figsize=(50, 20))
df_inter.plot(title="DataFrame Plot")
plt.legend(loc = 'upper right', fontsize=7, bbox_to_anchor= (1.4,1 )) 
plt.show()
```

```python
a = 0.125
y = f(t, a, b, c)
y_new = df_inter['f(0.125,b,c)']

plt.figure()
plt.plot(t, y, label='Real function f(x) for a = 0.125',  marker='o' ,color='red', markersize=5)
plt.plot(t, y_new, label='Interpolated points for a = 0.125', marker='o', color='blue', markersize=1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute the ratio
ratio = y/y_new

plt.figure(figsize=(15,8))
plt.plot(t, y, label='Real function scalated f(x)' ,color='red')
plt.scatter(t, ratio, label='Ratio between true values and interpolated values', marker='o', color='green', s= 8)
plt.legend()
plt.grid(True)
plt.show()
```
