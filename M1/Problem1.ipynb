```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Create a DataFrame with the dataset information
data = {
    'Dataset Name': [
        'Blog Authorship Corpus',
        'CIFAR-10',
        'UCF101',
        'UrbanSound8K'
    ],
    'Type': [
        'Language',
        'Image',
        'Video',
        'Audio'
    ],
    'Amount': [
        '140 million words',
        '60000 32x32 coloured images',
        '13320 videos',
        '8732 audios'
    ],
    'Size (GB)': [
        0.298,
        0.163,
        8.0,
        8.5
    ]
}

df = pd.DataFrame(data)
```

```python
#Make the plot
plt.figure(figsize=(10, 6))

#Create the bars so then we can put labels inside
bars = plt.barh(df['Dataset Name'], df['Size (GB)'], color=['blue', 'orange', 'green', 'red'])

plt.xlabel('Storage Size (GB)')
plt.title('Storage Needs of Different Datasets')


# Add labels to the bars
for bar, amount in zip(bars, df['Amount']):
    plt.text(bar.get_width() + 1.2, bar.get_y() + bar.get_height()/2, amount,
             ha='center', va='center', color='black', fontsize=10)

# Show the plot
plt.show()
```
